# Challenge 2 - Data Cleaning and Transformation

If your conclusion from the previous challenge is the data need cleaning/transformation, do it in the cells below. However, if your conclusion is the data need not be cleaned or transformed, feel free to skip this challenge. But if you do choose the latter, please provide rationale.

```python
# Select only numerical columns (excluding Channel & Region if they exist)
num_cols = df_wcd.select_dtypes(include=['int64', 'float64']).columns

# Compute Q1, Q3, and IQR for numerical columns
Q1 = df_wcd[num_cols].quantile(0.25)
Q3 = df_wcd[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Create a mask for detecting outliers
mask = ~((df_wcd[num_cols] < (Q1 - 1.5 * IQR)) | (df_wcd[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)

# Apply the mask to remove outliers
df_wcd = df_wcd[mask]

print(f"New dataset size after outlier removal: {df_wcd.shape}")
```

```python
# Convert boolean columns to integers (0 or 1)
df_wcd[['Channel_2', 'Region_2', 'Region_3']] = df_wcd[['Channel_2', 'Region_2', 'Region_3']].astype(int)

# Verify the change
print(df_wcd.dtypes)

df_wcd.head()
```

**Your comment here**

#### **Observations from Data Cleaning and Transformation**
1. **Outlier Removal using IQR:**
   - The dataset initially contained **440 rows**, which was reduced to **332 rows** after removing outliers.
   - The **Interquartile Range (IQR) method** was used to filter out values that were **1.5 times below Q1 or above Q3**.
   - This ensures that **extreme values do not distort clustering results**.

2. **Handling Boolean Columns:**
   - The categorical columns **`Channel_2`, `Region_2`, and `Region_3`** were initially in boolean format.
   - They were converted to **integer format (0 or 1)** to ensure compatibility with machine learning models.
   - This transformation maintains the categorical information while ensuring numerical processing compatibility.

3. **Potential Issues and Considerations:**
   - **Feature Scaling:** Since clustering models rely on **distance-based metrics**, the next step should be to apply **StandardScaler** to normalize numerical values.
   - **Impact of Outlier Removal:** The reduction of nearly **25% of the dataset** may affect model generalization, and further validation is needed.
   - **Feature Engineering:** Highly correlated variables may introduce redundancy, and one may be removed to improve clustering results.

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Import StandardScaler
from sklearn.preprocessing import StandardScaler

# Select only continuous numerical columns for scaling
num_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper','Delicassen']

# Initialize StandardScaler
scaler = StandardScaler()

# Fit and transform the data
customers_scale = scaler.fit_transform(df_wcd[num_cols])

# Convert back to a DataFrame for better readability
customers_scale = pd.DataFrame(customers_scale, columns=num_cols)

# Display the transformed data
print(customers_scale.head())
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# Your code here:
inertia = []
K_range = range(1, 11)  # Test cluster sizes from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customers_scale)
    inertia.append(kmeans.inertia_)

# Plot Elbow Method
plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia (WCSS)")
plt.title("Elbow Method for Optimal K")
plt.show()
```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
# Run K-Means with different cluster sizes

# Fit K-Means with the chosen optimal number of clusters
optimal_k = 2
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
customers_scale['Label'] = kmeans.fit_predict(customers_scale)

# Count cluster assignments
print(customers_scale['Label'].value_counts())
```

Count the values in `labels`.

```python
# Your code here:

from sklearn.metrics import silhouette_score

score = silhouette_score(customers_scale.iloc[:, :-1], customers_scale['Label'])
print(f"Silhouette Score for K-Means (k=3): {score:.4f}")
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
# Your code here:

# Experimenting with different numbers of clusters
cluster_range = [2, 3, 4, 5]
fig, axes = plt.subplots(1, len(cluster_range), figsize=(15, 5))

for i, k in enumerate(cluster_range):
    kmeans = KMeans(n_clusters=k, random_state=42)
    customers_scale[f'Label_{k}'] = kmeans.fit_predict(customers_scale)
    
    axes[i].scatter(customers_scale.iloc[:, 0], customers_scale.iloc[:, 1], c=customers_scale[f'Label_{k}'], cmap='viridis', alpha=0.6)
    axes[i].set_title(f'K-Means with {k} Clusters')

plt.show()
```

**Your comment here**

### **Observations on K-Means Clustering with Different Cluster Numbers**

#### **1. K=2 Clusters**
- The data is divided into two broad groups, **but the separation is not well-defined**.
- Some points that appear distinct are **forced into one of the two clusters**, potentially misrepresenting real groupings.
- This may work for **broad segmentation (e.g., high vs. low spenders)** but lacks granularity.

#### **2. K=3 Clusters**
- Introducing a third cluster helps **capture more variation in the data**.
- The separation appears **more meaningful**, with some points forming a new group that was previously lumped into the others.
- This may be a **better fit for segmenting customer types**.

#### **3. K=4 Clusters**
- The clusters start to **overlap**, but there is a more refined division among data points.
- Some **denser regions are further broken down**, but a few clusters seem close to each other.
- This may indicate that **K=4 is over-segmenting the data**, creating clusters that may not be distinctly meaningful.

#### **4. K=5 Clusters**
- More clusters are formed, **but some appear very close to each other**.
- Certain groups are **split unnecessarily**, leading to a **loss of clear separation**.
- The added complexity may **not contribute additional useful insights**.

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Your code here

# Experimenting with different DBSCAN parameters
eps_values = [0.5, 1.0, 1.5]
min_samples_values = [3, 5, 10]

fig, axes = plt.subplots(len(eps_values), len(min_samples_values), figsize=(15, 10))

for i, eps in enumerate(eps_values):
    for j, min_samples in enumerate(min_samples_values):
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(customers_scale)
        
        axes[i, j].scatter(customers_scale.iloc[:, 0], customers_scale.iloc[:, 1], c=labels, cmap='rainbow', alpha=0.6)
        axes[i, j].set_title(f'DBSCAN (eps={eps}, min_samples={min_samples})')

plt.show()
```

**Your comment here**

### **Observations on DBSCAN Clustering with Different `eps` and `min_samples` Values**

#### **1. Effect of `eps` (Neighborhood Radius)**
- **Small `eps` (0.5) results in many noise points** (purple dots not forming clear clusters).
- **Increasing `eps` (1.0, 1.5) results in more defined clusters**, as DBSCAN expands the reach of each core point.
- **At `eps=1.5`, clusters start merging**, potentially losing distinct group separations.

#### **2. Effect of `min_samples` (Minimum Points to Form a Cluster)**
- **Lower `min_samples` (3, 5) allows more clusters to form** but can lead to **more unstable, small clusters**.
- **Higher `min_samples` (10) forces clusters to be denser**, resulting in **fewer but more robust clusters**.
- When `min_samples` is too high, some points that should form a cluster remain unclassified (noise).

#### **3. Optimal Parameter Tradeoff**
| **Parameter Set** | **Observations** |
|------------------|----------------|
| **eps=0.5, min_samples=3**  | Too many outliers, not enough clusters. |
| **eps=0.5, min_samples=5**  | Some clusters form, but still many noise points. |
| **eps=0.5, min_samples=10**  | Only a few dense clusters form, many outliers. |
| **eps=1.0, min_samples=3**  | More balanced clusters, some noise remains. |
| **eps=1.0, min_samples=5**  | Clusters are more stable and distinct. |
| **eps=1.0, min_samples=10**  | Some cluster merging, but noise is reduced. |
| **eps=1.5, min_samples=3**  | Many points assigned to clusters, risk of over-grouping. |
| **eps=1.5, min_samples=5**  | Fewer clusters, some loss of resolution. |
| **eps=1.5, min_samples=10** | Over-merging of clusters, losing separation. |
